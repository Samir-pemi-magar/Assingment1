Task - 3 - Outlier Detection:
Tasks:
1. Identify outlier countries in South Asia based on their Score and GDP per Capita.
2. Define outliers using the 1.5 × IQR rule.

```python
# Calculate Q1 and Q3 for 'score' and 'Log GDP per capita'
Q1_score = df_south_asian['score'].quantile(0.25)
Q3_score = df_south_asian['score'].quantile(0.75)
IQR_score = Q3_score - Q1_score

Q1_GDP = df_south_asian['Log GDP per capita'].quantile(0.25)
Q3_GDP = df_south_asian['Log GDP per capita'].quantile(0.75)
IQR_GDP = Q3_GDP - Q1_GDP

# Calculate the lower and upper bounds for outliers for both 'score' and 'Log GDP per capita'
lower_bound_score = Q1_score - 1.5 * IQR_score
upper_bound_score = Q3_score + 1.5 * IQR_score

lower_bound_GDP = Q1_GDP - 1.5 * IQR_GDP
upper_bound_GDP = Q3_GDP + 1.5 * IQR_GDP

# Find outliers based on the conditions for both 'score' and 'Log GDP per capita'
outliers_iqr = df_south_asian[(df_south_asian['score'] < lower_bound_score) |
                              (df_south_asian['score'] > upper_bound_score) |
                              (df_south_asian['Log GDP per capita'] < lower_bound_GDP) |
                              (df_south_asian['Log GDP per capita'] > upper_bound_GDP)]

print(outliers_iqr)
```

3. Create a scatter plot with GDP per Capita on the x-axis and Score on the y-axis, highlighting outliers
in a different color.

```python
plt.figure(figsize=(10, 6))
plt.scatter(df_south_asian['Log GDP per capita'], df_south_asian['score'],
            color='blue', label='Normal', alpha=0.6)
plt.scatter(outliers_iqr['Log GDP per capita'], outliers_iqr['score'],
            color='red', label='Outliers', alpha=1.0)
plt.xlabel('Log GDP per capita')
plt.ylabel('Score')
plt.title('Scatter plot of GDP per Capita vs Score (Outliers Highlighted)')
plt.legend()
plt.show()
```
